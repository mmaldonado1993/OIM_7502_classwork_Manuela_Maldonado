```python
import matplotlib.pyplot as plt
region_counts = df['region'].value_counts(normalize=True) * 100  
region_counts.plot.bar() 
plt.xlabel("Region")
plt.ylabel("Proportion (%)")
plt.title("Regional Proportionality")

plt.show()
```
